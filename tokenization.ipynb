```python
nlp = spacy.blank("hi")

doc = nlp("सोचा था घंटे–दो घंटे में बंद हो जाएगी। यह सोचकर मैंने छाता भी अपने मित्र को लौटा दिया था। मेरे पास एक थैला था जिसमें केवल एक तौलिया थी। – जलमग्न जयपुर–गाड़ी जब जयपुर के प्लेटफार्म पर रुकी तो सुबह के चार बज चुके थे।")

for token in doc:
    print(token, token.is_currency, token.like_num, token.is_alpha)
```

```python
from spacy.symbols import ORTH

nlp.tokenizer.add_special_case("gimme", [
    {ORTH: "gim"},
    {ORTH: "me"}
])

doc = nlp("gimme double cheese extra large healthy pizza")
tokens = [token.text for token in doc]
tokens
```
